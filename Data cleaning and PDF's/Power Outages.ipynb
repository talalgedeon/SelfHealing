```python
climate=['Thunderstorms','Winds','Severe Weather','Wind','Storm','Cold','Ice','Snow','Ice/Snow','Tornadoes','Fires','Hailstorm','Lightning','Hurricane']

Vandalism=['Vandalism','Theft','Suspicious','Vandalism','Sabotage','Attack']

mechanical=['Generation Inadequacy','Transmission','Malfunction','Shed','Deficiency','Fault','Coal','Failure']

climate_change=[]
vandalism=[]
mechanical_fault=[]
other=[]

def categorize(x):
    for i in x:
        if x in climate:
            climate_change.append(x)
        elif x in Vandalism:
            vandalism.append(x)
        elif x in mechanical:
            mechanical_fault.append(x)
        else:
            other.append(x)
        return(climate_change, vandalism, mechanical_fault, other)  
            
            
            
    




```
